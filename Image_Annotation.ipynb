```python
import pandas as pd
from PIL import Image
```

```python
csv_file = 'supp_data01_new.csv'
df = pd.read_csv(csv_file)
df = df.drop(['Unnamed: 0'], axis=1)
display(df)
```

```python
for folder in df['set'].unique():
    classes_file_path = '/content/onedrive/trypoastigote_images/' + folder + '/classes.txt'

    with open(classes_file_path, 'a') as classes_file:
        classes_file.write("T_CRUZI")
        classes_file.close()
```

```python
tam_bbox = 100
```

```python
for i in df.index:
    image_file = '/content/onedrive/trypoastigote_images/' + df['set'][i] + '/' + df['filename'][i]
    image = Image.open(image_file)
    
    yolo_y = df['y'][i] / image.height
    yolo_y = str(yolo_y)
    
    yolo_x = df['x'][i] / image.width
    yolo_x = str(yolo_x)
    
    yolo_w = tam_bbox / image.width
    yolo_w = str(yolo_w)
    
    yolo_h = tam_bbox / image.height
    yolo_h = str(yolo_h)
    
    #yolo_class = [j for j, x in enumerate(classes) if x == df['structure'][i]][0]
    #yolo_class = str(yolo_class)
    yolo_class = "0"

    yolo_annotation_file_path = df['set'][i] + '/' + df['filename'][i]
    yolo_annotation_file_path = yolo_annotation_file_path.replace(".jpg", ".txt")
    
    yolo_str = yolo_class + ' ' + yolo_x + ' ' + yolo_y + ' ' + yolo_w + ' ' + yolo_h
    
    with open(yolo_annotation_file_path, 'a') as yolo_annotation_file:
        yolo_annotation_file.write(yolo_str + '\n')
        yolo_annotation_file.close()
```
